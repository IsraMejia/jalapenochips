### Diagrama Subrutinas

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grÃ¡fico dirigido
G = nx.DiGraph()

# Definir nodos y conexiones
nodes = [
    ("Inicio", "Nombres"),
    ("Nombres", "Jugadores"),
    ("Jugadores", "Inicio Juego"),
    ("Inicio Juego", "Nuevo Jugador"),
    ("Nuevo Jugador", "Ciclo Juego"),
    ("Ciclo Juego", "Si"),
    ("Si", "Ret1"),
    ("Si", "Ciclo Juego"),
    ("Ret1", "Ciclo Juego"),
    ("Ciclo Juego", "Fin Recorrido"),
    ("Fin Recorrido", "Nuevo Jugador"),
    ("Nuevo Jugador", "Fin Juego"),
    ("Fin Juego", "Ganador"),
    ("Ganador", "Obtener Minimo"),
    ("Obtener Minimo", "Ret1"),
    ("Ret1", "Ganador")
]

# AÃ±adir nodos y conexiones al grÃ¡fico
for edge in nodes:
    G.add_edge(*edge)

# Establecer posiciones de los nodos
pos = nx.fruchterman_reingold_layout(G)

# Dibujar el grÃ¡fico
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='skyblue', font_size=8, font_weight='bold', arrowsize=20)
plt.title("Diagrama de Flujo - Llamadas a Subrutinas")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Definir las formas y conexiones del diagrama de flujo
shapes = {
    "Inicio": (0, 0),
    "Obtener Minimo": (0, -1),
    "Ciclo Minimo": (0, -2),
    "Nuevo Minimo": (0, -3),
    "Sig1": (0, -4),
    "Fin": (0, -5)
}

connections = {
    "Inicio": ["Obtener Minimo"],
    "Obtener Minimo": ["Ciclo Minimo"],
    "Ciclo Minimo": ["Nuevo Minimo", "Sig1"],
    "Nuevo Minimo": ["Sig1"],
    "Sig1": ["Ciclo Minimo", "Fin"]
}

# Crear el diagrama de flujo
def draw_flowchart(shapes, connections):
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.axis('off')

    for shape, pos in shapes.items():
        ax.annotate(shape, pos, xytext=(-30, -15), textcoords='offset points',
                    arrowprops=dict(facecolor='black', arrowstyle='->'))

    for node, targets in connections.items():
        start = shapes[node]
        for target in targets:
            end = shapes[target]
            ax.plot([start[0], end[0]], [start[1], end[1]], 'k-')

    plt.show()

# Dibujar el diagrama de flujo
draw_flowchart(shapes, connections)
```

## Pseudocodigo CAD

```python
# FunciÃ³n para simular el proceso de conversiÃ³n A/D del PIC16F877(A)
def convertir_AD():
    # a. En el banco cero, limpiar el puerto A, CLRF PORTA
    PORTA = 0x00
    print(f"Puerto A despuÃ©s de limpiar: {PORTA:#04x}")
    
    # b. Cambiar al banco uno
    banco = 1
    print(f"Banco cambiado a: {banco}")
    
    # c. Definir al puerto A para entradas analÃ³gicas, escribir 00 al registro ADCON1
    ADCON1 = 0x00
    print(f"ADCON1 configurado: {ADCON1:#04x}")
    
    # d. Regresar al banco cero
    banco = 0
    print(f"Banco cambiado a: {banco}")
    
    # e. Realizar la configuraciÃ³n de la fuente de reloj, el canal de entrada y prender al convertidor A/D, en el registro ADCON0
    ADCON0 = 0b01000001  # Ejemplo: canal 0 seleccionado, A/D habilitado
    print(f"ADCON0 configurado: {ADCON0:#04x}")
    
    # f. Iniciar la conversiÃ³n colocando un â1â a la bandera GO/DONE# del registro ADCON0
    GO_DONE = 1
    ADCON0 |= (GO_DONE << 1)  # Set GO_DONE bit
    print(f"ConversiÃ³n iniciada, ADCON0: {ADCON0:#04x}")
    
    # g. Generar un retardo de 20 microsegundos para asegurar la adquisiciÃ³n de la tensiÃ³n de entrada
    import time
    time.sleep(20e-6)  # 20 microsegundos
    
    # h. Esperar a que GO/DONE# sea igual a cero, lo que indica que ha concluido el proceso de conversiÃ³n
    # En un microcontrolador real, esto se verifica constantemente hasta que el bit se borre
    # AquÃ­ simularemos una espera breve
    GO_DONE = 0
    ADCON0 &= ~(1 << 1)  # Clear GO_DONE bit
    print(f"ConversiÃ³n completada, ADCON0: {ADCON0:#04x}")
    
    # i. Leer el resultado de la conversiÃ³n del registro ADRESH
    ADRESH = 0xAB  # Valor simulado para el resultado de la conversiÃ³n
    print(f"Resultado de la conversiÃ³n: {ADRESH:#04x}")

# Llamar a la funciÃ³n para ejecutar el proceso de conversiÃ³n A/D
convertir_AD()
```
